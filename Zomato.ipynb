```python
import pandas as pd

df=pd.read_csv('C:/Users/SAHNAWAJ/Desktop/python data/zomato_dataset.csv')

df.head()
```

```python
## Summary statistics of the dataset
print(df.describe())
```

```python
# Checking for missing values
Null_Value=(df.isnull().sum())
print(Null_Value)
```

```python
# Filter restaurants with a dining rating greater than 4.5
high_rated_restaurants = df[(df['Dining_Rating'] > 4.5)]
 

Ligh_rated_restaurants = df[(df['Dining_Rating'] > 4.5)]

#high_rated_restaurants
Ligh_rated_restaurants
```

```python
# Filter restaurants with delivery rating greater than 4.0
Resturant_with_delivery_rating=df[(df['Delivery_Rating'] > 4.0)]

Resturant_with_delivery_rating
```

```python
# Filter restaurants with a dining rating greater than 4.5 and delivery rating greater than 4.0
high_rate_restaurants = df[(df['Dining_Rating'] > 4.5) & (df['Delivery_Rating'] > 4.0)]
print(high_rate_restaurants)
```

```python
# Calculate the total dining votes for each restaurant in descending vote order
total_dining_votes = df.groupby('Restaurant_Name')['Dining Votes'].sum()
sorted_data=total_dining_votes.sort_values(ascending=False)
print(sorted_data)
```

```python
# Calculate the average prices of food items for each City
average_prices_by_city = df.groupby('City').agg(mean_price=('Prices','mean'))
average_prices_by_city = average_prices_by_city.reset_index()
print(average_prices_by_city)
```

```python
# Calculate the total dining votes for each restaurant ascending order
total_dining_votes = df.groupby('Restaurant_Name')['Dining Votes'].sum()
sort_data=total_dining_votes.sort_values(ascending=True)
print(sort_data)
```

```python
# Top 10 Restaurants with the Highest Dining Votes
top_dining_restaurants = df.groupby('Restaurant_Name')['Dining Votes'].sum().nlargest(10)

print(top_dining_restaurants)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
average_prices_by_city.plot(kind="bar")
plt.xlabel('City')
plt.ylabel('Average Price')
plt.title('Average Prices in Different Cities')
plt.show()
```

```python
# Average Prices in Different Cities in bar chart
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
sns.barplot(x='City', y='mean_price', data=average_prices_by_city)
plt.xticks(rotation=45)
plt.xlabel('City')
plt.ylabel('Average Price')
plt.title('Average Prices in Different Cities')
plt.show()
```
